```python
import duckdb

df = duckdb.sql("SELECT * FROM '../data/mart/state_party_contributions.parquet' ORDER BY contribution_received_date").df()
df.head()
```

```python
import plotly.express as px

fig = px.line(df, x="contribution_received_date", y="total_raised_cumulative", title='Cumulative Raised By Caucus', color='party', facet_col='branch')
fig.show()
```

```python
fig = px.line(df, x="contribution_received_date", y="total_raised_cumulative_year", title='Year Cumulative Raised By Caucus', color='party', facet_col='branch')
fig.show()
```

```python
year_grouped_df = duckdb.sql("""
    SELECT
        organization,
        party,
        branch,
        donation_received_year,
        SUM(total_raised) AS total_raised
    FROM '../data/mart/state_party_contributions.parquet'
    GROUP BY ALL
    ORDER BY donation_received_year, party
""").df()
year_grouped_df.head(10)
```

```python
fig = px.bar(year_grouped_df, x="donation_received_year", y="total_raised", title='Total Raised By Caucus By Year', color='party', facet_col='branch', barmode='group')
fig.show()
```
